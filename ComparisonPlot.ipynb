```python
import pickle
import numpy as np
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
L=[[2**-i,0] for i in range(1,11)]
```

```python
result=pd.DataFrame(columns=['trainloss','testloss','epoch','algorithm','seed','eta','momentum'])
eta,momentum=[2**-i for i in range(10) , 0]
for eta, momentum  in L:
    for experiment in range(10):
        logfile = "logs/signSGDm_seed_%d_lr_%f_momentum_%f.log"%(experiment, eta, momentum)
        print logfile
        if os.path.exists(logfile):
            data = pickle.load(open(logfile,'rb'))
            temp=pd.DataFrame(data)
            temp.columns=['trainloss','testloss','epoch','algorithm','seed','eta','momentum']
            result=result.append(temp)
            print result
```
